```python
dense_index = pc.Index(index_name)
```

```python
dense_index.upsert_records("example-namespace", records)
```

```python
dense_index.describe_index_stats()
```

```python
query = "Famous historical structures and monuments"
```

```python
results = dense_index.search(
    namespace="example-namespace",
    query={"top_k": 10, "inputs": {"text": query}},
)
```

```python
for hit in results["result"]["hits"]:
    print(
        f"id: {hit['_id']:<5} | score: {round(hit['_score'], 2)} | "
        f"{hit['fields']['category']:<10} | text {hit['fields']['chunk_text']:<50}"
    )
```

```python
reranked_results = dense_index.search(
    namespace="example-namespace",
    query={"top_k": 10, "inputs": {"text": query}},
    rerank={"model": "bge-reranker-v2-m3", "top_n": 10, "rank_fields": ["chunk_text"]},
)
```

```python
for hit in reranked_results["result"]["hits"]:
    print(
        f"id: {hit['_id']:>5},",
        f"score: {round(hit['_score'], 2):>5}",
        f"category: {hit['fields']['category']:>10}",
        f"text: {hit['fields']['chunk_text']:<50}",
    )
```
